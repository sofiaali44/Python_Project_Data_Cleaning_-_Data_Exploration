## Key Insights

### 1. United Kingdom is the top most generating revenue for the Online Retail Store.
- United kingdom is the leading country generation up to $8.2 million. That is 8 times more than the the second-leading-country Netherland which is generation $0.3 million.
- The results suggest that the business relies heavily on the UK market, with significantly lower revenue contributions from other countries.

### 2. Monthly sales trend is increases as the months go by

- Monthly sales were relatively low at the beginning of the year and generally increased as the year progressed, reaching their highest level in November.

- This pattern suggests increased customer demand toward the end of the year, likely influenced by seasonal shopping and holiday purchases.

### 3. Dotcom Postage was the highest revenue-generating product.

- **Dotcom Postage** was the highest revenue-generating product, followed closely by the **Regency Cakestand 3 Tier**.
- These products contributed significantly more revenue than the other top-selling products, making them key drivers of the store's overall sales.

### 4. A high percentage of missing `CustomerID` values limits customer-level analysis.

- Approximately **33.33%** of the `CustomerID` values are missing, making it difficult to accurately identify the store's most valuable customers.
- Transactions with missing `CustomerID` account for approximately **$1.4 million** in annual revenue.
- As a result, customer purchasing behavior, customer lifetime value, and the highest-value customers cannot be determined accurately for these transactions.

### 5. World War 2 Gliders Asstd Designs was the most purchased product.

- **World War 2 Gliders Asstd Designs** had the highest number of units purchased.
- The top five most purchased products have relatively similar purchase volumes, indicating only a small difference in demand between them.
- This suggests that customers consistently purchase a variety of products, rather than showing a strong preference for a single item.
