```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# Loading dataset

df = pd.read_csv(r"C:\Users\User\Desktop\week 1&2\2_Netflix_Dataset_Cleaning_& _EDA\netflix_titles.csv")
```

```python
# copy dataset for cleaning

df1 = df.copy()
```

```python
# identify missing values
df1.isna().sum()
```

```python
# missing value % of total

round((df1.isna().sum()/df1.count())*100,)
```

```python
# handing missing values with fill na

df1["director"] = df1["director"].fillna("unknown")
df1["cast"] = df1["cast"].fillna("unknown")
df1["country"] = df1["country"].fillna("unknown")
```

```python
df1.isna().sum()
```

```python
# Check duplicates while ignoring the primary key column

df1.duplicated(subset=df1.columns.drop("show_id")).sum()
```

```python
# date formatting
df.info()
```

```python
# fixing the date column dtype
df1["date_added"] = df1["date_added"].str.strip()

df1["date_added"] = pd.to_datetime(
    df1["date_added"],
    format="%B %d, %Y"
)
```

```python
df1.info()
```

```python
# Select text columns that need cleaning
text_cols = ["type", "title", "director", "cast", "country", "rating","duration"]

# Remove spaces and standardize text formatting
for col in text_cols:
    df1[col] = (
        df1[col]
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )

# Fix capitalization for categorical columns
category_cols = ["type", "country", "listed_in"]

for col in category_cols:
    df1[col] = df1[col].str.title()
```

```python
# cleaning rating columns

df["rating"].unique()
```

```python
# Move duration values from rating to duration
df1.loc[df1["rating"].str.contains("min", na=False), "duration"] = df1["rating"]

# Replace incorrect ratings with missing values
df1.loc[df1["rating"].str.contains("min", na=False), "rating"] = np.nan
```

```python
# saving the cleaned dataset
df1.to_csv("netflix_dataset_cleaned.csv", index=False)
```

## Issues Found and Actions Taken

| **Issue Found** | **Action Taken** |
|-----------------|------------------|
| **date_added** was stored as an object data type. | Used `pd.to_datetime()` to convert the column to a datetime data type. |
| Missing values were found in the **director**, **cast**, **country**, and **date_added** columns. | Replaced missing values in **director**, **cast**, and **country** with `"Unknown"`. The missing values in **date_added** were left unchanged because only 10 records were affected. |
| No duplicate values were found. | No action was required. |
| Three values in the **rating** column were incorrectly placed and belonged in the **duration** column. | Moved the misplaced values to the **duration** column and replaced the corresponding **rating** values with missing values. |
