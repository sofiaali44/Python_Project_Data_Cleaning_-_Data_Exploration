# Key Insights

### 1. Netflix has significantly more movies than TV shows

The Netflix dataset contains **6,131 movies and 2,676 TV shows**. Movies are more than **twice the number of TV shows**, showing that Netflix's content library has a much stronger focus on movies compared to TV series.

### 2. The United States dominates Netflix content production

The United States is the leading content-producing country in the dataset, with a **large gap between itself and the second-highest producing country**. The US produces approximately **three times more content than India**, highlighting the strong dominance of American-produced titles in Netflix's catalog.

### 3. Netflix content additions increased rapidly between 2014 and 2019

The number of titles added to Netflix grew significantly between 2014 and 2019, reaching its highest point around 2019. This period represents a major expansion of Netflix's content library.

### 4. TV-MA is the most common content rating on Netflix

TV-MA is the most frequent rating in the dataset, with **3,207 titles**, followed by TV-14 with **2,160 titles**. This shows that a large portion of Netflix's content is targeted toward mature audiences, with fewer titles aimed at younger viewers under the age of 13.

### 5. International Movies and Dramas are among the most common genres on Netflix

The most frequent genres in the Netflix dataset are **International Movies and Dramas**, indicating that Netflix has a strong focus on global movie content and drama-based titles. This suggests that Netflix's content strategy includes a wide variety of international and drama productions to appeal to a global audience.
